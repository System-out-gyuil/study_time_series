```python
import pandas as pd
import json

with open('./datasets/bitcoin_2010_2024.json') as f:
    json_data = json.load(f)

bitcoin_df = pd.DataFrame(json_data['market-price'])
bitcoin_df
```

```python
from datetime import datetime

def changeDate(milis):
    timestamp = milis / 1000
    convert_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return convert_time
```

```python
bitcoin_df.x = bitcoin_df.x.apply(changeDate)
bitcoin_df
```

```python
bitcoin_df.set_index('x', inplace=True)
bitcoin_df
```

```python
bitcoin_df.info()
```

- index가 datetime이 아니라서 시계열 데이터로 변환

```python
bitcoin_df.index = pd.to_datetime(bitcoin_df.index)
bitcoin_df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
plt.plot(bitcoin_df[bitcoin_df.index > '2016'])
plt.show()
```

```python
pre_bitcoin = bitcoin_df[bitcoin_df.index > '2016']
pre_bitcoin
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(1, 2, figsize=(12, 8))

plot_acf(pre_bitcoin, lags=20, ax=ax[0])
plot_pacf(pre_bitcoin, lags=20, ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

plot_acf(pre_bitcoin.diff().dropna(), lags=20, ax=ax[0])
plot_pacf(pre_bitcoin.diff().dropna(), lags=20, ax=ax[1])
plt.show()
```

```python
pre_bitcoin_diff = pre_bitcoin.diff().dropna()
pre_bitcoin_diff
```

```python
window_size = 20

moving_avg = pre_bitcoin.rolling(window_size).mean()
print(moving_avg)

moving_avg_diff = pre_bitcoin_diff.rolling(window_size).mean()
print(moving_avg_diff)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 20))

plt.subplot(2, 1, 1)
plt.plot(pre_bitcoin, c='orange')

plt.subplot(2, 1, 1)
plt.plot(moving_avg)

plt.title('Google stock MR (2014~2024)')

plt.subplot(2, 1, 2)
plt.plot(pre_bitcoin_diff, c='orange')

plt.subplot(2, 1, 2)
plt.plot(moving_avg_diff)

plt.title('Google stock differencing MR (2014~2024)')

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd

model = ARIMA(pre_bitcoin, order=(1, 2, 1))
model_fit = model.fit()

start_index = pd.to_datetime('2021-01-04')
end_index = pd.to_datetime('2024-03-21')
forecast = model_fit.predict(start=start_index, end=end_index)

plt.figure(figsize=(15, 8))
plt.plot(pre_bitcoin, label='original')
plt.plot(forecast, label='predicted', c='orange')
plt.title("Time Series Forecast")
plt.legend()
plt.show()
```
